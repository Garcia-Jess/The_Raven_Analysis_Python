# Edgar Allan Poe's "The Raven" Analysis Using Python

#### "The Raven" is a poem written by Edgar Allan Poe and was first published in 1845.


Reference: https://www.poetryfoundation.org/poems/48860/the-raven


```python
import nltk
import googletrans
```

```python
# Load the text of "The Raven"
with open("The_Raven.txt", "r", encoding="utf-8") as file:
    text = file.read()
print("The total number of characters, including spaces and punctuation: ", len(text))
```

### 1. Text Preprocessing:

#### Preprocess the text by removing punctuation and stopwords, converting all words to lowercase, and splitting the text into tokens (words).

```python
from nltk.tokenize import word_tokenize
import string
from nltk.corpus import stopwords

# Tokenize the text into words
words = word_tokenize(text)

# Remove punctuation tokens
words_without_punctuation = [word.lower() for word in words if word.isalnum()]

# Get English stopwords
english_stopwords = set(stopwords.words('english'))

# Remove stopwords from the tokenized words list
words_without_stopwords = [word for word in words_without_punctuation if word.lower() not in english_stopwords]

print(words_without_stopwords)
```

## 2. Frequency Distribution

#### Calculate the frequency distribution of words in the text.

```python
from nltk.probability import FreqDist

# Calculate the frequency distribution of words
fdist = FreqDist(words_without_stopwords)

# Print the most common words
print(fdist.most_common(10))
```

```python
import matplotlib.pyplot as plt

# Plot the frequency distribution
fdist.plot(15, cumulative=False)
plt.show()
```

## 3. Sentiment Analysis:

#### Analyze the sentiment of the text

```python
from nltk.sentiment import SentimentIntensityAnalyzer


# Initialize the Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()

# Compute the sentiment score for the entire poem
sentiment_score = sia.polarity_scores(text)

# Print the sentiment score
print(sentiment_score)
```

```python
# Extract sentiment scores
sentiment_scores = {k: v for k, v in sentiment_score.items() if k != 'compound'}

# Plot the sentiment scores
plt.bar(sentiment_scores.keys(), sentiment_scores.values())
plt.xlabel('Sentiment')
plt.ylabel('Score')
plt.title('Sentiment Analysis of the Poem')
plt.show()
```

#### X-axis: The x-axis represents the sentiment categories: 'neg' for negative sentiment, 'neu' for neutral sentiment, and 'pos' for positive sentiment. These categories correspond to the sentiment dimensions analyzed by the SentimentIntensityAnalyzer.

#### Y-axis: The y-axis represents the sentiment scores ranging from 0 to 1. A score closer to 0 indicates a weaker sentiment, while a score closer to 1 indicates a stronger sentiment.

#### Compound: The compound sentiment score is a summary score that represents the overall sentiment of the text. It's a combination of the negative, neutral, and positive scores.
